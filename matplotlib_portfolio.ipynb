```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset(name='mpg')
df.head()
```

```python
plt.scatter(x= df.weight, y=df.mpg, marker= '*')
```

En este grfico podemos inferir la relacin lineal que existe entre el consumo de gasolina y el peso del motor. A mayor peso menor distancia recorrida.

```python
plt.hist(x=df.mpg, bins = 40, edgecolor= 'black')
```

```python
freq = df.origin.value_counts()
freq.index
```

```python
freq.values
```

```python
plt.bar(x=freq.index, height=freq.values, color=['blue', 'green', 'red'])
```

En este grfico podemos inferir que la mayor cantidad de vehculos producidos son originarios de Estados Unidos.

```python
sns.scatterplot(x='weight', y='mpg', data= df, edgecolors='black', hue='origin')
plt.xlabel('weight')
plt.ylabel('mpg')
```

En este grfico podemos inferir que cada punto representa un coche, a mayor peso el auto recorre menos distancia.

```python
import plotly.express as px
px.scatter(x='weight', y='mpg', data_frame=df, color='origin', hover_name=df.name)
```
